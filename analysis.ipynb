# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load all sheets from the excel workbook

```python
all_sheets = pd.read_excel('Datanova data updated.xlsx', sheet_name=None)
```

Inspecting the sheets

```python
print(all_sheets.keys())
```

```python
eda_sheets = {name: df for name, df in all_sheets.items() if not name.lower().startswith('sheet')}
```

```python
print(eda_sheets.keys())
```

```python
for name, sheet in eda_sheets.items():
    print(f"\n=== Sheet: {name} ===")
    print("Shape:", sheet.shape)
    print("Columns:", sheet.columns.tolist())
    print("Data types:\n", sheet.dtypes)
    print("Missing values:\n", sheet.isnull().sum())
    print("First 5 rows:\n", sheet.head())
    print("Descriptive statistics:\n", sheet.describe(include='all'))
    print("-" * 40)
```

# ECCDE - Early Child Care Development E

## 1. Teacher Distribution & Workforce Composition

```python
df_eccde_teacher_qual = eda_sheets['Quaified o not quaified teacher']
```

```python
df_eccde_teacher_qual.head()
```

```python
print(df_eccde_teacher_qual.columns)
```

### 1.1 Qualified vs. Unqualified Teachers by Region and School Type

```python
plot_df = df_eccde_teacher_qual[['state', 'qual_teacher_total', 'non_qual_teacher_total']].copy()
plot_df.set_index('state', inplace=True)
plot_df.plot(kind='bar', figsize=(12, 6))
plt.title('Qualified vs Non-Qualified Teachers by State')
plt.xlabel('State')
plt.ylabel('Number of Teachers')
plt.xticks(rotation=45)
plt.legend(['Qualified', 'Non-Qualified'])
plt.tight_layout()
plt.show()
```

### 1.2 Teacher-to-Pupil Ratio (PTR) by State and Level

```python

ptr_df = eda_sheets['pupilteacher rtio'][['state', 'total']].copy()
ptr_df.set_index('state', inplace=True)

ptr_df.plot(kind='bar', figsize=(14, 6))
plt.title('Teacher-to-Pupil Ratio (PTR) by State')
plt.ylabel('PTR (Pupils per Teacher)')
plt.xlabel('State')
plt.xticks(rotation=45)
plt.legend(['Pupils per teacher'])
plt.tight_layout()
plt.show()
```

## 2. Professional Qualification & Continuous Development

## 2.1 TRCN Registration Rate by State

```python
trcn_df = eda_sheets['no f trcn staff'][['state', 'registered_total', 'unregistered_total']].copy()
trcn_df['registered_total'] = pd.to_numeric(trcn_df['registered_total'], errors='coerce')
trcn_df['unregistered_total'] = pd.to_numeric(trcn_df['unregistered_total'], errors='coerce')
trcn_df.set_index('state', inplace=True)
# renaming crossriver stae value
trcn_df.index = trcn_df.index.str.strip().str.replace('\r', '').str.replace('\n', '')
trcn_df.rename(index={'CROSS-_x000D_RIVER': 'CROSS_RIVER'}, inplace=True)
print(trcn_df.index.unique())
trcn_df = trcn_df[trcn_df.index != 'NATIONAL']
```

```python
trcn_df.plot(kind='bar', figsize=(14, 6))
plt.title('TRCN Registered vs Unregistered Teachers by State')
plt.ylabel('Number of Teachers')
plt.xlabel('State')
plt.xticks(rotation=45)
plt.legend(['Registered', 'Unregistered'])
plt.tight_layout()
plt.show()
```

### 2.2 CPD Participation Rate

```python
cpd_df = eda_sheets['no of cpdstaff']
```

```python
cpd_teacf_df = cpd_df[['state', 'lv_1_total','lv_2_total', 'lv_3_total', 'lv_4_total', 'lv_5_total', 'none_total']].copy()
cpd_teacf_df.set_index('state', inplace=True)
cpd_teacf_df.plot(kind='bar', figsize=(14, 6))
plt.title('Levels of teachers in CPD by state')
plt.xlabel('State')
plt.ylabel('CPD Levels of Teachers')
plt.xticks(rotation=45)
plt.legend(['Level 1', 'Level 2', 'Level 3', 'Level 4', 'Level 5', 'None'])
plt.tight_layout()
plt.show()
```

## 3. Digital Competence & ICT Readiness

### 3.1 Teachers with ICT Training or Computer Literacy

```python
df_ict = all_sheets['digital literacy'][['state', 'literate_total', 'not_literate_total']].copy()
df_ict.set_index('state', inplace=True)
df_ict.plot(kind='bar', figsize=(14, 6))
plt.title('Teachers digital literacy by state')
plt.xlabel('State')
plt.ylabel('Digital Literacy of Teachers')
plt.xticks(rotation=45)
plt.legend(['literate_total', 'not_literate_total'])
plt.tight_layout()
plt.show()
```
